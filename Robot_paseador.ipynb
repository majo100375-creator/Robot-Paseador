```python
import time
import random

class RobotPaseador:
    def __init__(self, id_robot):
        self.id = id_robot
        self.bateria = random.randint(20, 100)
        self.ubicacion = "Central"
        self.bolsas = 10
        self.deposito_desechos = []
        self.en_servicio = False

    def verificar_estado(self):
        print(f"\n[Robot {self.id}] Verificando estado...")
        # Lógica de batería
        if self.bateria < 50:
            print(f"Batería al {self.bateria}%. Cargando...")
            self.bateria = 100
        else:
            print(f"Batería optimizada al {self.bateria}%.")

        # Verificar suministros
        if self.bolsas > 0:
            print("Suministros: OK.")
            return True
        return False

    def reaccionar_entorno(self, evento):
        if evento == "charco":
            print("¡Charco detectado! Calculando ruta para evitarlo.")
        elif evento == "defecar":
            print("El perro ha defecado. Iniciando protocolo de limpieza...")
            self.recoger_desechos()
        elif evento == "otro_perro":
            print("Contacto con otro perro detectado. Manteniendo calma y siguiendo ruta.")
        elif evento == "lluvia":
            print("Está lloviendo. Buscando refugio inmediato...")
            time.sleep(2)
            print("Resguardado. Esperando a que pare...")
            print("Lluvia cesó. Reanudando servicio.")

    def recoger_desechos(self):
        print("1. Sacando dispensador de bolsas.")
        print("2. Tomando bolsa.")
        print("3. Inclinándose y recogiendo desechos.")
        print("4. Cerrando bolsa y guardando en depósito.")
        self.bolsas -= 1
        self.deposito_desechos.append("Desecho recogido")

    def navegar(self, destino):
        print(f"Navegando hacia {destino} siempre por la vereda (Evitando la calle).")

class AppServicio:
    def reservar_servicio(self):
        print("--- Reserva del Servicio ---")
        nombre_perro = input("Nombre de la mascota: ")
        tamano = input("Tipo de perro (Grande, Mediano, Chico): ")
        tipo_servicio = input("Tipo de servicio (Tiempo): ")
        distancia = input("Por favor ingrese en metros (distancia): ")
        domicilio = input("Dirección de recogida: ")
        print(f"\nServicio confirmado para {nombre_perro}. Asignando robot...")
        return {"mascota": nombre_perro, "domicilio": domicilio, "tipo": tipo_servicio}

    def post_servicio_pago(self):
        print("\n--- Pago y Calificación ---")
        pago = input("Seleccione método de pago: ")
        propina = input("Ingrese propina: ")
        print("Procesando pago... ¡Éxito!")

        print("\nEncuesta de satisfacción:")
        input("1. ¿Qué tan satisfecho estás? ")
        input("2. ¿Cómo está el perro? ")
        input("3. ¿Mejorarías algo? ")
        print("¡Gracias por tu feedback!")
```

```python
import time
import random

class RobotPaseador:
    def __init__(self, id_robot):
        self.id = id_robot
        self.bateria = 100
        self.bolsas = 10
        self.nivel_agua = 100  # Nuevo: Nivel de agua en %
        self.historial_paseos = [] # Nuevo: Historial de servicios

        # Base de datos integrada: Seguridad Vial y Normativa Madrid
        self.base_datos_madrid = {
            "seguridad_vial": ["Caminar por la vereda", "Prioridad peatón", "Cruce por paso de cebra"],
            "normas_mascotas": [
                "Uso obligatorio de correa",
                "Recogida obligatoria de excrementos",
                "Identificación microchip (REIAC)",
                "Horarios parques (verano/invierno)"
            ]
        }

    def verificar_suministros(self):
        print(f"\n[Check de Sensores] Robot {self.id}:")
        # Sensor de bolsas
        if self.bolsas < 2:
            print("⚠️ Alerta Sensor: Pocas bolsas. Reponer en central.")
        else:
            print(f"✅ Sensor de bolsas: {self.bolsas} disponibles.")

        # Sensor de agua
        if self.nivel_agua < 20:
            print("⚠️ Alerta Sensor: Nivel de agua bajo.")
        else:
            print(f"✅ Dispensador de agua: {self.nivel_agua}% capacidad.")

    def ofrecer_agua(self):
        print("Protocolo Hidratación: Desplegando dispensador de agua.")
        self.nivel_agua -= 5
        print("Mascota hidratada.")

    def registrar_paseo(self, mascota, duracion, incidentes):
        entrada = {
            "fecha": time.strftime("%Y-%m-%d %H:%M"),
            "mascota": mascota,
            "duracion": duracion,
            "incidentes": incidentes
        }
        self.historial_paseos.append(entrada)
        print(f"Servicio registrado en la base de datos para {mascota}.")

    def cumplir_normativa(self):
        print("\n[Consultando Base de Datos Madrid]")
        for regla in self.base_datos_madrid["normas_mascotas"]:
            print(f"Aplicando regla: {regla}")

# --- Simulación de la mejora ---
bot = RobotPaseador("MAD-BOT-01")
bot.verificar_suministros()
bot.cumplir_normativa()
bot.ofrecer_agua()
bot.registrar_paseo("Rex", "45 min", "Lluvia leve, resguardado en portal")

print("\nHistorial Guardado:", bot.historial_paseos)
```

```python
def ejecutar_paseo():
    app = AppServicio()
    bot = RobotPaseador("RB-01")

    # 1. Reserva
    datos_reserva = app.reservar_servicio()

    # 2. Preparación
    if bot.verificar_estado():
        bot.navegar(datos_reserva['domicilio'])

        # 3. Llegada y Recogida
        print(f"\n[Llegada] Robot toca el timbre en {datos_reserva['domicilio']}.")
        print("Robot: '¡Hola! He venido por el paseo de " + datos_reserva['mascota'] + "'.")
        print("Humano coloca la correa en el arnés de seguridad.")

        # 4. El Paseo (Simulación de eventos)
        print("\n--- Iniciando Ruta ---")
        bot.en_servicio = True
        eventos_camino = ["charco", "otro_perro", "defecar", "lluvia"]

        for e in eventos_camino:
            time.sleep(1)
            bot.reaccionar_entorno(e)

        # 5. Devolución
        print(f"\n[Regreso] Entregando a {datos_reserva['mascota']} en domicilio.")
        print("Humano retira al perro. Confirmación en App mediante deslizamiento.")
        print("Robot: '¡Hasta luego! Ha sido un placer'.")

        # 6. Post-Servicio
        bot.navegar("Central")
        print("Robot en central. Conectando a base de carga...")
        bot.bateria = 100

        app.post_servicio_pago()

# Ejecutar el programa
if __name__ == "__main__":
    ejecutar_paseo()
```
