```python
import time

class RobotPaseador:
    def __init__(self, modelo):
        self.modelo = modelo
        self.bateria = 45  # Simulación de estado inicial
        self.ubicacion = "Central"
        self.deposito_bolsas = 0
        self.normas_madrid = []

    def configurar_sistema(self):
        print(f"\n--- [⚙️ Configuración {self.modelo}] ---")

        # Lógica de batería
        if self.bateria < 50:
            print(f"¡Hola! Mi batería está al {self.bateria}%. Me cargaré un poquito hasta el 100% para dar el mejor paseo.")
            self.bateria = 100
        else:
            print("¡Energía a tope! Estoy listo para comenzar esta aventura.")

        print("Sincronizando la normativa de Madrid con mucho cuidado...")
        self.normas_madrid = ["Limpiar orina con agua", "Recoger excrementos", "Uso de correa"]
        print("Rutas listas y equipo de limpieza preparado. ¡Todo en orden!")

    def hidratar_mascota(self, mascota, minuto):
        print(f"\n[💧 HIDRATACIÓN]: Han pasado {minuto} minutos. Me detengo suavemente para que {mascota} beba agua fresca.")
        print(f"[📸 APP NOTIFICACIÓN]: Enviando foto de {mascota} bebiendo a la App del dueño... ¡Se ve muy feliz!")

    def reaccionar_a_entorno(self, evento):
        if evento == "charco":
            print("[SISTEMA]: ¡Uy, un charco! Recalculando nuestra ruta para que no se moje las patitas.")
        elif evento == "perro_defeca":
            print("[SISTEMA]: ¡Oh! Hora de cumplir con mi deber de limpieza.")
            print("-> Con mucho cuidado: saco bolsa, recojo, cierro y guardo para mantener limpia la ciudad.")
            self.deposito_bolsas += 1
        elif evento == "lluvia":
            print("[SISTEMA]: ¡Están cayendo gotas! Buscando un refugio seco y seguro de inmediato.")
            time.sleep(2)
            print("[SISTEMA]: Ya paró de llover. ¡Podemos seguir disfrutando del aire libre!")
        elif evento == "pis":
            print("[SISTEMA]: Aplicando agua con jabón según la normativa de Madrid. ¡Impecable!")

    def caminar(self):
        print("-> Caminando alegremente por la acera. Mantengo a tu mascota lejos del tráfico por seguridad.")

class AppPaseo:
    def __init__(self):
        self.planes = {
            "Bronce": {"min": 15, "mtrs": 350, "precio": 15},
            "Plata": {"min": 30, "mtrs": 750, "precio": 30},
            "Oro": {"min": 60, "mtrs": 1000, "precio": 45}
        }

    def gestionar_reserva(self):
        print("\n--- [📱 App BotPro: Registro] ---")
        email = input("Email: ")
        password = input("Contraseña: ")
        nombre_dueño = input("Nombre del dueño: ")
        apellido = input("Apellido: ")
        dni = input("DNI: ")
        input("¿Acepta términos de protección de datos? (S/N): ")

        print("\n--- [📱 Reserva de Paseo] ---")
        mascota = input("Nombre de la mascota: ")
        chip = input("Código Microchip: ")
        fecha = input("Fecha del paseo: ")
        hora = input("Hora de recogida: ")
        tamano = input("Tipo de perro (Grande/Mediano/Pequeño): ")

        print("Planes disponibles: Bronce (15€), Plata (30€), Oro (45€)")
        plan_sel = input("Seleccione plan: ").capitalize()

        total = self.planes[plan_sel]["precio"]
        minutos_totales = self.planes[plan_sel]["min"]

        extra = input("¿Deseas tiempo extra? (0.50€/min) (si/no): ").lower()
        if extra == "si":
            minutos_extra = int(input("Ingresa minutos extras: "))
            total += (minutos_extra * 0.50)
            minutos_totales += minutos_extra

        indicaciones = input("Indicaciones especiales (dejar en blanco si no hay): ")
        domicilio = input("Introduce datos de domicilio: ")

        return {
            "dueño": nombre_dueño,
            "mascota": mascota,
            "total": total,
            "domicilio": domicilio,
            "minutos": minutos_totales,
            "codigo_recogida": "7744",
            "codigo_entrega": "1234" # Código de seguridad para la llegada
        }

# --- EJECUCIÓN DEL PROCESO ---

# 1. Reserva y Configuración
app = AppPaseo()
reserva = app.gestionar_reserva()

robot = RobotPaseador("BotPro v2.1")
robot.configurar_sistema()

# 2. Inicio del Servicio
print(f"\n--- [🚀 Inicio del Servicio] ---")
print(f"Robot llega a {reserva['domicilio']}. *Toca el timbre con una melodía amigable*")
print(f"Robot: '¡Hola, {reserva['dueño']}! Qué alegría verte. He venido para llevar a {reserva['mascota']} a divertirse.'")

cod_ingresado = input(f"Humano, por favor ingresa el código de seguridad ({reserva['codigo_recogida']}): ")
if cod_ingresado == reserva['codigo_recogida']:
    print("¡Perfecto! Arnés desbloqueado y ajustado con suavidad. ¡Estamos listos!")
    input("Humano: Confirma inicio de paseo en la App (Presiona Enter)")
else:
    print("Vaya, el código no coincide. Por seguridad de la mascota, no puedo iniciar el servicio.")
    exit()

# 3. El Paseo con Hidratación
print("\n--- [🐕 En el Paseo] ---")
robot.caminar()
robot.reaccionar_a_entorno("charco")

# Simulación de hidratación cada 10 minutos
for m in range(10, reserva['minutos'] + 1, 10):
    robot.hidratar_mascota(reserva['mascota'], m)

robot.reaccionar_a_entorno("perro_defeca")
robot.reaccionar_a_entorno("pis")
robot.reaccionar_a_entorno("lluvia")
print(f"Robot: '¡Qué buen paseo! {reserva['mascota']} se ha portado genial. Volvemos a casa.'")

# 4. Devolución y Desbloqueo de Seguridad
print("\n--- [🏠 Devolución y Seguridad] ---")
print(f"He llegado al domicilio de {reserva['mascota']}. El arnés permanece BLOQUEADO por seguridad.")

# NUEVA LÓGICA DE DESBLOQUEO AL LLEGAR
cod_final = input(f"Dueño, por seguridad ingresa el código en tu App para desbloquear el arnés ({reserva['codigo_entrega']}): ")
if cod_final == reserva['codigo_entrega']:
    print(f"¡Código correcto! Arnés liberado. Ha sido un placer cuidar de {reserva['mascota']}.")
    input("Deslice el botón en la app para confirmar recepción (Enter)")
    print("Robot: '¡Adiós! Espero que nos veamos pronto para otra aventura'.")
else:
    print("Código incorrecto. No puedo liberar el arnés. Llamando a soporte técnico por seguridad...")
    exit()

# 5. Post-Servicio Robot
print("\n--- [🔧 Mantenimiento Central] ---")
print(f"Vaciando depósito de forma ecológica: {robot.deposito_bolsas} desechos procesados.")
robot.bateria = 100
print("Estoy cargado, limpio y muy feliz por el trabajo de hoy. ¡Listo para el siguiente!")

# 6. Post-Servicio Humano (Pago y Feedback)
print("\n--- [💳 Pago y Calificación] ---")
print(f"Total a pagar por este maravilloso paseo: {reserva['total']}€")
metodo_pago = input("Seleccione método de pago: ")
propina = input("¿Deseas dejar una pequeña propina para mi mantenimiento? ")
print("Procesando el pago... ¡Todo perfecto!")

# Calificación
satis = input("Del 1 al 10, ¿te ha gustado mi compañía hoy?: ")
puntual = input("¿Qué tal me porté con la puntualidad? (1-10): ")
sugiere = input("¿Tienes alguna sugerencia para que mejore? (si/no): ").lower()
if sugiere == "si":
    input("Cuéntame, te escucho con atención: ")

print(f"\n{reserva['dueño']}, mil gracias por confiar en BotPro. {reserva['mascota']} y yo te deseamos un día fantástico.")
```
